This notebook shows an example of how to retrain 21cmVAE and the AuteoEncoder-based Emulator (from Appendix A).

```python
%load_ext autoreload
%autoreload 2
```

```python
import VeryAccurateEmulator as VAE
from matplotlib import pyplot as plt
```

## Training 21cmVAE

```python
emulator = VAE.VeryAccurateEmulator()
emulator.emulator.summary()
```

```python
emulator.train()
```

```python
fig = plt.figure()
ax = fig.add_axes([0., 0., 1., 1.])
ax.plot(emulator.train_losses, label='Training')
ax.plot(emulator.val_losses, label='Validation')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_yscale('log')
plt.show()
```

```python
rmse = emulator.compute_rms_error()
print(rmse.mean()*100)
```

```python
import numpy as np
errs = np.empty((20, 1704))
for i in range(20):
    print('i = ', i)
    emulator = VAE.VeryAccurateEmulator()
    emulator.train()
    rmse = emulator.compute_rms_error()
    errs[i] = rmse
    if i == 0:
        print('Saving first emulator')
        emulator.emulator.save('best_em_nov8_0.h5')
    elif np.mean(rmse) < np.mean(errs[:i], axis=1).min():
        print('Saving new best emulator: ', i)
        fname = 'best_em_nov8_' + str(i) +'.h5'
        emulator.emulator.save(fname)
np.save('nov8_manyems.npy', errs)
print(np.mean(errs, axis=1).argmin())
print(np.mean(errs, axis=1).min())
```

## Training the Autoencoder-based Emulator

```python
ae_emulator = VAE.AutoEncoderEmulator()
# ae_emulator.autoencoder.summary()
# ae_emulator.emulator.summary()
# ae_emulator.encoder.summary()
# ae_emulator.decoder.summary()
```

```python
ae_emulator.train()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0., 0., 1., 1.])
ax1.plot(ae_emulator.ae_train_losses, label='Training')
ax1.plot(ae_emulator.ae_val_losses, label='Validation')
ax1.legend()
ax1.set_title('AutoEncoder')
ax1.set_ylabel('Loss')
ax2 = fig.add_axes([1.1, 0., 1., 1.])
ax2.plot(ae_emulator.em_train_losses, label='Training')
ax2.plot(ae_emulator.em_val_losses, label='Validation')
ax2.legend()
ax2.set_title('Emulator')
plt.setp([ax1, ax2], xlabel='Epoch', yscale='log')
plt.show()
```

```python
rmse = ae_emulator.compute_rms_error()
print(rmse.mean()*100)
```

```python
import numpy as np
errs = np.empty((20, 1704))
for i in range(20):
    print('i = ', i)
    ae_emulator = VAE.AutoEncoderEmulator()
    ae_emulator.train()
    rmse = ae_emulator.compute_rms_error()
    errs[i] = rmse
    if i == 0:
        print('Saving first emulator')
        ae_emulator.emulator.save('best_ae_em_nov8_0.h5')
        ae_emulator.autoencoder.save('best_autoencoder_nov8_0.h5')
        ae_emulator.decoder.save('best_decoder_nov8_0.h5')
        ae_emulator.encoder.save('best_enocder_nov8_0.h5')
    elif np.mean(rmse) < np.mean(errs[:i], axis=1).min():
        print('Saving new best emulator: ', i)
        ae_emulator.emulator.save('best_ae_em_nov8_'+str(i)+'.h5')
        ae_emulator.autoencoder.save('best_autoencoder_nov8_'+str(i)'.h5')
        ae_emulator.decoder.save('best_decoder_nov8_0'+str(i)+'.h5')
        ae_emulator.encoder.save('best_enocder_nov8_0'+str(i)+'.h5')
```
