## Training a new model

```python
custom_model = VeryAccurateEmulator.VeryAccurateEmulator() # instantiate the VAE class
```

```python
latent_dimension = 10
encoder_dimensions = [128, 256]
decoder_dimensions = [214]
emulator_dimensions = [64]
beta = 1
gamma = 0.5
custom_model.set_hyperparameters(latent_dim=latent_dimension, encoder_dims=encoder_dimensions,
                                      decoder_dims=decoder_dimensions, em_dims=emulator_dimensions,
                                      beta=beta, gamma=gamma)
print('Hyperparameters are set to:')
print('Latent dimension:', custom_model.latent_dim)
print('Encoder dimensions:', custom_model.encoder_dims)
print('Decoder dimensions:', custom_model.decoder_dims)
print('Emulator dimensions:', custom_model.em_dims)
print('Beta:', beta)
print('Gamma:', gamma)
```

```python
custom_model.train()
```
